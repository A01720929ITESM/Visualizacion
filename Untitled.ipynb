```python
df_r = df.rename(columns={'Sex ID':'sex_id','Sex':'sex','Age Range':'age_range','Population':'population','Percentage':'percentage'})
```

```python
df_r.columns
```

```python
df_male= df_r[0:18]
```

```python
df_male
```

```python
df_female= df_r[18:36]
```

```python
df_female
```

```python
y = range(0, len(df_male))
x_male = df_male['population']
x_female = df_female['population']
```

```python
fig, axes = plt.subplots(ncols=2, sharey=True, figsize=(9, 6))
fig.patch.set_facecolor('xkcd:light grey')
plt.figtext(.5,.95,"Population Pyramid in San Pedro ", fontsize=15, ha='center')
axes[0].barh(y, x_male, align='center', color='royalblue')
axes[0].set(title='Males')
axes[1].barh(y, x_female, align='center', color='lightpink')
axes[1].set(title='Females')
axes[1].grid()
axes[0].set(yticks=y, yticklabels=df_male['age_range'])
axes[0].invert_xaxis()
axes[0].grid()
plt.show()
```

```python
df_g = df_r
df_group = df_g.groupby('age_range').sum()
df_group.head(30)
```

```python
df_group.pop('sex_id')
```

```python
df_group.head()
```

```python
df_group.columns
```

```python
df_group = df_group.reset_index()
df_group.head()
```

```python
plt.figure(figsize=(40,10))
sns.set(font_scale=1.5)
sns.regplot(
    x='age_range',
    y='percentage',
    data=df_group,
    fit_reg=False
)
plt.title('Population vs Percentage 2020',fontsize=30)
plt.xlabel('Age Range',fontsize=25)
plt.ylabel('Percentage',fontsize=25)
plt.show()
```
