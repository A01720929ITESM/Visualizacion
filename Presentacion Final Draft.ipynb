```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df= pd.read_csv('datasets/covid-sanpedro.csv')
df.head(37)
```

```python
df.pop('Time ID')
```

```python
df.pop('Age Range ID')
```

```python
totalc = sns.boxplot(data=df.loc[:,['Total_Cases']])
plt.title('Casos Totales de Covid')
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.set(font_scale=1.5)
sns.regplot(
    x='Sex',
    y='Total_Cases',
    data=df,
    fit_reg=False
)
plt.title('Casos Hombres vs Mujeres')
plt.show()
```
